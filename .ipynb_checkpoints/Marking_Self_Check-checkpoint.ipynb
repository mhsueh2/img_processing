```python
from IPython.display import clear_output as __clear_output
        
def __compress_correct(__img_ch, n_components=5):
    __patches = patchify(__img_ch)
    __pca = PCA(n_components=5)
    __pca.fit(__patches)
    
    __means, __components, __img_size = __pca.mean_, __pca.components_, __img_ch.shape
    __compressed_data = np.dot(__patches - __means, __components.T)
    __comp_img = {'y': __compressed_data, 'aux_data': (__components, __means, __img_size)}
    return __comp_img

def __decompress_correct(__comp_data):
    __y = __comp_data['y']
    __components, __means, __img_size = __comp_data['aux_data']
    __decompressed_channel = np.dot(__y, __components) + __means
    __img_ch = depatchify(__decompressed_channel, img_size=__img_size)
    
    return __img_ch


def __compress_image_correct(__img):
    __img_cf = convert_to_cf(__img)
    return [__compress_correct(__img_cf[i]) for i in range(3)]
    

def __decompress_image_correct(__comp_img):
    return np.dstack([__decompress_correct(__comp_img[i]) for i in range(3)])



def __expected(__val, __exp, __e=1e-2):
    return np.mean(np.abs(__val - __exp)) < __e


def __run_notebook(notebook):
    try:
        %run "$notebook"
    except Exception as __err:
        print(__err)
        return False
    
    return True


def __run_task(__task, *args):
    
    __marks, __result = __task(*args)
    return (__marks, __result)
```
